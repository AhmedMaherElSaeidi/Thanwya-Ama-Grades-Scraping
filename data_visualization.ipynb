### Visualization part

```python
# Pie chart
df_gender_grades = df_grades[['gender', 'total']].groupby('gender', axis=0).sum()
df_gender_grades['total'].plot.pie(autopct='%1.01f%%', title='total_grade_among_genders')
plt.legend(labels=['Female', 'Male'], loc='upper right')
```

```python
# Pie chart
plt.figure(figsize=(9,9))
df_gender_grades = df_grades[['section', 'total']].groupby('section', axis=0).sum()
df_gender_grades['total'].plot.pie(autopct='%1.01f%%', title='total_grade_among_sections')

sections = df_grades['section'].unique()
plt.legend(labels=sections, loc='upper right')
```

```python
# Box plot
array_of_1_categories = ['passed']
array_of_0_categories = ['absence', 'postponed', 'cancelled', 'exempt', np.nan]

df_grades_values = df_grades.iloc[:, 7:].replace(array_of_0_categories, 0).replace(array_of_1_categories, 1).astype(float)
df_grades_values.head()
```

```python
plt.figure(figsize=(10,20))
for index, column in enumerate(df_grades_values.columns):
    plt.subplot(6,3,index + 1)
    df_grades_values[column].plot.box()
```

```python
#
plt.figure(figsize=(10, 10))
df_gender_grades = df_grades[['school_id', 'total']].groupby('school_id', axis=0).sum().nlargest(15, 'total')
df_gender_grades['total'].plot.bar(title='total_grade_among_schools', rot=45)
```
